```python
import pandas as pd
import matplotlib.pyplot as plt
```

My data set is Marvel movies

```python
data = { 
    'Movie': ['Iron Man', 'The Avengers', 'Black Panther', 'Captain Marvel'],
    'Year': [2008, 2012, 2018, 2019],
    'Budget': [140, 220, 200, 152],
    'Box Office': [585, 1519, 1347, 1128],
    'Rotten Tomatoes Score': [94, 91, 96, 79]
}

df = pd.DataFrame(data)

print(df.describe())
```

My code for the visuals of the data

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Movie'], df['Box Office'], color='skyblue')
plt.title('Box Office Earnings for Marvel Movies')
plt.xlabel('Movie')
plt.ylabel('Box Office (Million USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
